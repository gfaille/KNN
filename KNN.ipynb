```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
from sklearn.neighbors import KNeighborsClassifier

```

```python
tableau = pd.read_csv("data.csv", sep=",")
print(tableau)
```

```python
tableau.shape
tableau.head(10)
```

```python
tableau.describe()
```

```python
tableau.plot(x='sepal_length', y='sepal_width', kind="scatter")
```

```python
sns.scatterplot(tableau, x='sepal_length', y='sepal_width', hue='species')
sns.set_theme()
```

```python
sns.scatterplot(tableau, x='petal_length', y='petal_width', hue='species')
sns.set_theme()
```

```python
X = tableau[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
Y = tableau['species']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)

knn = KNeighborsClassifier(n_neighbors= 10)
knn.fit(X_train, Y_train)
knn.score(X_test, Y_test)
```

```python
prediction_espece = knn.predict([[6, 3, 4, 1]])
prediction_espece[0]
```
